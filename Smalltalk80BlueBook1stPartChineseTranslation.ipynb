```python
'''
19 
数字字面量也可以被表达在一个无非十进制基于前于那数字带着一个半径前缀。
那半径前缀包括那值于那数字半径（总是用十进制表达）被跟着那字母“r“。
下面的例子指定数字在八进制带着它们对应的十进制值。

当那基数大于10，那大写首字母开始带着 A 被使用为了数字大于9。
下面的例子指定数目在十六进制带着它们相应地十进制值。

20
数字字面量也可以被表达在一个科学记数通过在值的数字后带着一个指数后缀。
那指数后缀包括字母 e 跟着的被那指数（以十进制表达）。
那数字指定之前于娜指数后缀是倍数由那半径升到那次方由那指数指定的。

 字符
字符是对象其代表那单独的符号于一个字母。一个字符字面量表达式组成于一个美元符号跟着任何字符，例如，
$a
$M
$-
$$
$1

 字符串
字符串是对象其代表一系列的字符。字符串相应于消息其访问单独字符，替代下属字符串，执行比较与其他字符串。那字面量代表于一个字符串是一个系列于字符被界定于单引号，例如，
’hi'
'food'
'the Smalltalk-80 system'
任何字符可能被包括在一个字符串字面量。如果一个单引号被包含在一个字符串之内，它必须被双写去避免混淆与那界定。例如，那字符串字面量
‘can''t'
 指向一个字符串于那五个字符$c,$a,$n,$',和$t.

 符号
符号是对象其代表字符串被用作名字在那系统。那字面量代表于一个符号是一个序列于字母字符前于被一个英镑符号，例如，
#bill
#M63

不会有两个符号用那同样地字符；每一个符号都是独一无二的。这使高效的比较符号成为可能。

数组
一个数组是一个简单地数据结构对象其内容可以被参考通过一个整数索引从1到一个数目其是那大小于那数组。 数组相应消息其请求方位到它们的内容。
那字面量代表于一个数组是一系列其他字面量--数目，字符，字符串，符号，和数组--被界定由括号和前置一个英镑符号。
其它字面量被空格分开。内嵌的符号和数组不被前置英镑符号。
一个数组于三个数目被描述由那表达式
#(1 2 3)
一个数组于7个字符串被描述由表达式
#（'food' 'utilities' 'rent' 'household' 'transportation' 'taxes' 'recreation')
一个数组于两个数组和两个数目被描述由表达式
#((’one' 1) ('not' 'negative') 0 -1)
 一个数组于一个数，一个字符串，一个符号，和其他数组被描述由表达式
#(9 'nine' $9 nine (0 'zero' $0 () 'e' $f 'g' $h 'i'))

 变量
一个对象的可访问记忆由变量组成。
大部分变量有名称。
每一个变量记忆一个单独的对象且那变量名可以被用作一个表达式去访问到那对象。
从一个特定地点可以访问的对象被决定由哪一个变量名是可达到的。
例如，

22
内容于一个对象的实例变量是不可访问的对于其它对象因为那些变量的名称可以被使用只在那方法于那对象的类。

一个变量名是一个简单地识别码，一个系列于字母和数字开始于一个字母。 一些例子于变量名是：
index
initialIndex
textEditor
bin14 
bin14Total
HouseholdFinances
Rectangle
IncomeReasons
有两种变量在这个系统，区别于如何广泛他被被访问。
私有变量只可以被访问由一个单独的对象。
实例变量是私有的。
共享变量可以被访问由多于一个对象。
私有变量名被要求必须用小写首字母。
共享的变量名被要求必须用大写首字母。
头五个例子识别码标示符展示在上面指向私有变量和那后面三个指向共享变量。

其它大写管理明显的在例子中上面是标示符由若干单词组成的要大写每一个单词首字母从第一个之后。这个管理不是强制的在系统中。

赋值
一个字面量常量总是指向同一个对象，但是一个变量名可能指向不同的对象在不同时间。
变量名指向的对象会被更改当一个赋值表达式被求值。
赋值早期不算一种表达式因为任何表达式可以变成一个赋值通过包含一个前缀赋值语句。
一个赋值语句前缀被构成由名于那变量其值将被改变跟着一个左箭头<-.
下面的例子是一个字面值表达式其有一个赋值前缀。
它指示那变量名为quantity的这是将指向对象其代表数值19。
quantity<- 19
下面的例子是一个变量名表达式带着一个赋值前缀。
它决定那变量名名为 index 应该指向同一个对象如同名为 initialIndex 的变量。

23
index <- initialIndex
 其它例子于赋值表达式是：
chapterName := 'Expression Syntax'
flavers := #('vanilla' 'chocolate' 'butter pecan' 'garlic')
 多于一个的赋值前缀可以被包括，决定那值于几个变量的被改变。
index := initailIndex := 1
这个表达式决定了两个变量名为 index 和 initialIndex 应该指向数字1.
消息表达式和块表达式也可以有赋值前缀，如下的章节会看到。

伪变量名
一个伪变量名是一个标识符其指向一个对象。在这个方向，它类似于一个变量名。
一个伪变量名是不同于一个变量名是它的值不可以被改变用一个赋值表达式。
一些伪变量在系统中是常量；它们总是指向同一个对象。
三个重要伪变量名字是 nil，true，和 false。

nil 指向一个对象被用作那值于一个变量当没有其他对象是合适的。
 变量没有被初始化的指向 nil。

true 指向一个对象其代表逻辑正确。
它是被用作一个肯定的响应到一个消息做一个建大的真假查询。

false 指向一个对象其代表逻辑错误。
它被用作一个否定的响应到一个消息做一个简单的真假查询。

那对象名为 ture 和 false 被叫做布林对象。
它们代表那答案到真假问题。
例如，一个数字将响应带着 true 或者 false 到一个消息询问是否那数字是大于另外一个数。
布林对象响应到消息其计算逻辑功能和执行条件控制结构。

其它伪变量在系统(如，self 和 super）其值是不同的依赖于它们被用在哪里。
下面三章将会讨论它们。

24
消息
消息代表那接卖弄之间于微言-80的组件。
 一个消息要求一个操作在那接收者方面。
一些列子于消息表达式和那互动它们代表如下。

消息到数字表带表数学操作。
3+4 计算3和四的和。
index+1 向名为 index 的数加一。
index>limit 查询是否名为 index 的树大于名为 limit 的数。
theta sin 计算名为 theta 的数的正弦值。
quantity sqrt 计算名为 quantity 的数的正平方根。

消息到线性数据结构代表那添加或者移除于信息。


list addFirst: newComponent
添加名为 newComponent 的对象作为首个元素于娜线性数据结构名为 list。

list removeLast
 移除并返回 list 的最后一个元素。

消息到结对的数据结构（如一个字典）代表那添加或者移除于信息。
ages at: 'Brett Jorgense' put: 3
绑定字符串‘Brett Jorgensen’ 与 数字3在名为 ages 的字典里。

address at：’Peggy Jorgensen‘
查看对象结对于字符创’Peggy Jorgenson‘ 在名为 addresses 的字典。

消息到矩形代表图形查询和计算。
frame center
 回答名为 frame 的矩形的中心点得位置。

frame containsPoint：cursorLocation
 回答真如果名为 cursorLocation 的位置内在于名为 frame 的矩形内，否则回答 false。

25
frame intersect：clippingBox
计算代表名为 frame 和名为 clippingBox 的两个矩形的交际矩形。

消息到财务历史记录代表交易和查询。
HouseholdFinances spend: 32.50 on：’utilities‘
通知那财务历史名为 HouseholdFinances 的$32.50已经花在了水电公用设置账单。

HouseholdFinances totalSpentFor：’food‘
询问 HouseholdFinances 多少钱已经被花在了食物上。

选择器和参数

一个消息表达式描述一个接收者，选择器，和可能一些参数。
接收者和参数是被描述由其他表达式。
选择器被指定字面上的。

一个消息的选择器是一个名字为了互动的种类那发送者要求接收者的。
例如，在消息
 theta sin
 接收者是一个数字被指向 由那名为 theta 的变量名且选择器是 sin。
接收者决定如何去相应消息（此例中，如何去计算那正弦值于它的值）。
在两个消息表达式
3+4
和
 previousTotal + increment
 那选择器是+。
两个消息询问那接收者去计算和返回一个和。
这些消息每一个都包含一个对象作为添加到那选择器（4在第一个表达式和 incement 在第二个。）
额外的对象在那消息里是参数其指定被加入的数量。

如下的俩个消息表达式描写了同类型的操作。
接收者是一个 FinancialHistory 的实例且将会返回花在特定原因的钱数。
参数决定兴趣的原因。
第一个表达式要求那为了支付水电公共费用的花费数量。
HouseholdFinances totalSpentOn：’utilities‘
一个消息的选择器决定接收者的哪种操作会被召唤。
参数是其它对象其被卷入的在那被选定的操作。

一元消息 消息布袋参数被叫做一元消息。
例如，手上现金根据 HouseholdFinances 是那值于那一元消息表达式
 HouseholdFinances cashOnHand
 这些消息被叫做一元消息因为只有一个对象，那接收者，被卷入。
一个一元消息选择器可以是任何简单标识符。
其它一元消息表达式例子是
 theta sin
quantity sqrt
nameString size

关键字消息
带着一个或者多个参数的消息类型是关键字消息。
关键字消息的选择器由一个或者多个关键字组成，在每一个参数之前。
一个关键字是一个简单的标示符带着一个尾冒号。
例子于表达式描述单关键字消息是
 HouseholdFinances totalSpentOn: 'utilities'
index max: limit
 一个消息带着两个参数的将有一个选择器带着两个关键字。
例子于表达式描述双关键字消息是
 HouseholdFinances spend：30.45 on: 'food'
ages at: 'Brett Jorgensen' put:3
当那选择器于一个多关键字消息被指向独立的，那些关键字是被串联的。
最后两个消息的选择器是 spend：on：和 at：put：。可以由任意数目的关键字在一个消息中，但是大部分消息在系统中有少于三个。

p27
二元消息
还有另一种欧冠消息表达式带着一个单独参数，叫做二元消息。
一个二元消息的选择器被一个或者两个非字母数字的字符组成。
唯一个约束是第二个字符不可以是一个减法符号。
二元选择器意在被用作数学消息。
列子于二元消息表达式是
3+4
total-1
total<= max

 返回值
微言-80消息提供双向交流。
 选择器和参数提交信息到接收者关于什么类型的响应去运作。
 接收者提交信息回去通过返回一个对象其变成那值于那消息表达式。
如果一个消息表达式包括一个赋值前缀，那对象被接收者返回的将变成那变量指向的新的对象。
例如，表达式
 sum ：=3+4
使得7称为那名为 sum 的变量的新值。表达式
 x：=theta sin
 使得那正弦值于 theta 称为名为 x 的变量的新值。
如果 theta 的值是1，那么 x 的新值就变成了0.841471.
如果 theta 的值是1.5，那么 x 的新值就变成了0.997495.

index 指向的数目可以被增加通过表达式
 index：=index+1
即使没有信息需要被交流回那发送者，一个接收者总是返回一个值对于那消息表达式。
返回一个值决定着对消息的响应已经完成。
一些消息意味着只是去通知那接收者某些事情。
列子是那消息去记录财务交易被描述由那如下的例子。
HouseholdFinances spend：32.50 on: 'utilities'
HouseholdFinances receive: 1000 from: 'pay'
 这些接收者于这些消息通知那发送者只是这是一个完成的记录关于那交易。
默认值被返回的在这样的例子是通常那接收者自己。
如表达式
 var：=HouseholdFinances spend：32.50 on：‘utilities’
导致 var 指向同一个财务历史如 HouseholdFinances。

解析
所有那消息表达式展示至今为止已经描述了那接收者和参数带着字面量或者变量名。
当接收者或变量名于一个消息表达式被描述由其他消息表达式，那议题于表达式是如何被解析的就升起了。
一个例子于一个一元消息描述那接收者于另一个一元消息是
1.5 tan rounded
 一元消息被解析从左到右。
例子中第一个消息是那一元选择器 tan 被送往1.5.
这个表达式的值（约14.1014）接收那一元消息 rounded 且返回那最近的整数，14.
树14是整个表达式的值。

二元消息也是被解析从左到右。
一个例子于二元消息描述那接收者于另一个二元消息是
 index+offset*2
由 index 从消息+offset 返回的值是二元消息*2的接收者。

所有二元选择器都有同样地优先级；只有那顺序在其它们被写得问题。
主意这使得数学表达式在微言-80语言不同于那些在其他语言其乘法和除法优先于加法和减法。

括号可以被使用去改变求值的顺序。
一个消息内在于括号被发送之前于任何消息外在于那括号。
如果那前一个例子被写成
 index +（offset*2）
那乘法将被执行之前于加法。

一元消息优先级高于二元消息。
如果一元消息和二元消息同时出现，一元消息将会被首先发送。
在例子

p29
frame width+border width*2
frame width 的值是接收者于那二元消息其选择器是+和那些参数是border width 的值。
+消息的值是，在返回中，那接受者于那二元消息*2.
这个表达式被解析如同它被括号了的如下：
（（frame width）+（border width））*2
括号可以被用作发送二元消息之前于一元消息。表达式
2*theta sin
 计算 theta 正弦值的两倍，当表达式
（2*theta）sin
计算两倍 theta 之后的正弦值。

无论关键字出现在一个没有括号的消息，它们组成一个单一的选择器。
因为这个串联，那里没有从左到右的解析规则对于关键字消息。
如果一个关键字消息被用作一个接收者或者参数于其他关键字消息，它必须被括号扩住。
表达式
 frame scale: factor max:5
描述一个单一的两个参数的关键字消息其选择器是 scale:max:。表达式
 frame scale：（factor max：5）
描述两个单一关键字消息其选择器是 scale:  和 max：
表达式 factor mac：5的值是参数为了那 scale：消息到 frame。

二元消息优先级高于关键字消息，
当一元消息，二元消息，和关键字消息出现在同一个表达式而没有括号，一元消息被发送首先，二元其次，关键字消息最后。例如
 bigFrame width：smallFrame width*2
被求值就像它被括号了得如下：
bigFrame width：（（smallFrame width）*2）

p30
 在下面的例子，一个一元消息描述那接收者于一个关键字消息和一个二元消息描述了那参数。
OrderedCollection new add：value*rate
总结解析规则：
1.一元表达式从左到右解析。
2.二元表达式从左到右解析。
3.二元表达式优先级高于关键字表达式，
4.一元表达式优先级高于二元消息。
5.被括号的表达式优先级高于一元表达式。

格式惯例
一个程序员是自由的去格式化表达式在不同的方式使用空格，tabs，和回车。
例如，多关键字消息经常被写带着每一个关键字参数配合于一个不同的行，如
 ages at：‘Brett Jorgensen'
      put: 3
或者
 HouseholdFinances
      spend：30.45
      on:'food'
 唯一的时候一个空格，制表符，或者回车影响一个表达式的意思是当它的消失将导致俩个字母或者俩个数目挤在一起。

叠
有一个特别的语法形式叫做叠其指定多个消息到同一个对象。
任何顺序的消息可以被表达而不用叠。
然而，叠经常降低使用变量的需求量。
一个层叠的消息表达式组成由一个接收者的描述跟着若干消息其被分开由分号。
例如
OrderedCollection new add：1；add：2；add：3
三个 add：消息被发送到那结果于 OrderedCollection new。
没有层叠，这将需要四个表达式和一个变量。

p31
例如，下面的四个表达式，由点号分开，有同样地结果如那层叠的表达式上面。
temp：=OrderedCollection new.
temp add: 1.
temp add: 2.
temp add: 3.

块
块是对象被用在许多控制结构在微言-80系统。
一个块代表一个被延迟的行动系列。
一个块表达式组成于一系列表达式被点号分开且被方括号界定。
例如，
[index := index+1]
或者
[index := index+1.
 array at: index put: 0]
如果一个点号跟着最后的表达式，它被忽略，如在
[expenditures at: reason.]
当一个块表达式被遭遇，那陈述封装在那括号不被立刻执行。
一个块表达式的值是一个对象其可以执行这些被封装的表达式在一个稍微晚点的时候，
当要求这样做得时候。
例如，表达式
 actions at: 'monthly payments'
         put: [HouseholdFinances spend: 650 on: 'rent'.
               HouseholdFinances spend: 7.25 on: 'newspaper'.
               HouseholdFinances spend: 225.50 on: 'car payment']
并没有实际发送任何 spend：on：消息到 HouseholdFinances。
它简单的绑定一个块跟那字符串‘monthly payments’。

一个块所描述的一系列动作将发生当那块收到那一元消息 value。
例如，如下两个表达式由同一的效果。
p32
index := index+1
和
[index := index + 1] value
被如下表达式指向的对象
 actions at：‘monthly payments’
 是那块包含三个 spend：on：消息。
表达式的执行
（actions at：‘monthly payments’）value
 导致那三个 spend：on：消息被送往 HouseholdFinances。
一个块也可以被赋值到一个变量。
所以如果表达式
 incrementBlock ：=[index := index + 1]
被执行，那么表达式
 incrementBlock value
 增加 index
当一个 value 消息被发送到一个块其返回的对象是那一些列表达式中最后一个的值。
所以如果表达式
 addBlock:=[index + 1]
被执行，这时另一种方法去增加 index 是去求值
 index：= addBlock value
 一个块其包含没有表达式返回 nil 当发送消息 value。
表达式
[]value
 有值 nil

 控制结构
一个控制结构决定顺序于一些行为。
基础控制结构在微言-80语言提供一些类表达式将被有序的求值。
许多非顺序的控制结构可以被实施用块。
这些控制结构被召唤通过发送一个消息到一个块或者通过发送一个消息带着一个或者多个块作为参数。
响应对于一个这种控制结构的消息决定那行为的顺序带着那 value 形式的消息被发送到那块。

测试如下表达式序列的求值会给出一个例子于块工作的方式。
incrementBlock：=[index := index +1].
sumBlock := [sum+ (index*index)].
sum:=0.
index:=1.
sum:=sumBlock value.
incrementBlock value.
sum:= sumBlock value
15个行为发生作为结果于求值这一些列表达式
1.赋值一个块到 incrementBlock。
2.赋值一个块到 sumBlock。
3.赋值数0到 sum。
4.赋值数1到 index。
5.发送消息 value 到块 sumBlock。
6.发送消息*1到数字1.
7.发送消息+1到数字0.
8.赋值数1到 sum。
9.发送消息 value 到块 **IncrementBlock.
10. 发送消息+1到数1.
11.赋值数2到 index。
12.发送消息 value 到块 sumBlock。
13.发送消息*2到数2.
14.发送消息+4到数1.
15.赋值数5到 sum。

一个例子于孔子结构由块实施的是简单重复，由一个被发送到一个整数带着 timesRepeat：作为选择器和一个块作为参数的消息来表达。
整数将相应而发送如它自身值决定的数目的 value 消息到那块。
例如，如下的表达式双倍那名为 amount 的变量的值4次。
4 timesRepeat：[amount ：= amount + amount]

p34
 条件
两个普通控制结构用块实施的时条件选择和条件重复。
条件选择类似于如果-那么-否则语句在 Algol 类似的语言和条件重复类似 while 和 until 陈述在那些语言。
这些条件控制结构使用那两个布林对象名为真和假被描述在伪变量的章节。
布林被返回从消息其询问简答的对错问题（如，那大小比较消息：<,=,<=,>,>=,~=).

条件选择
条件选择于一个行为被提供通过一个消息到一个布林带着选择器 ifTure：ifFalse：和两个块作为参数。
唯一的对象其懂得 ifTrue：ifFalse：消息的时真和假。
它们由相反的响应：真发送 value 到那第一个参数块和忽略第二个；假发送 value 到第二个参数块和忽略第一个。
例如，如下表达式赋值0或者1到名为 parity 的变量决定于是否变量 number 的值可被2整除。
那二元消息\\计算那整数或余数功能。
(number \\ 2)=0
     ifTrue:[parity := 0]
     ifFalse:[parity:=1]
ifTrue:ifFalse:返回值是那被执行的块的值。
前面的例子也可以被写为：
parity := (number \\ 2) = 0 ifTrue:[0] ifFalse: [1]
另外于ifTrue:ifFalse:有两个单关键字的消息其指定只一个条件结果。
那选择器于这些消息是 ifTrue：和 ifFalse：。
这些消息由同样地效果如那 ifTrue：ifFalse：消息当一个参数是一个空块。
例如，这两个表达式有同样地效果。
index <= limit
    ifTrue:[total := total + (list at: index)]
和
index <= limit
    ifTrue:[total := total + (list at: index)]
    ifFalse:[]
因为那值于一个空块是 nil ，那下面的表达式将设 lastElement 到 nil 如果 index 是大于 limit 的。

p35
lastElement:= index > limit ifFalse: [list at: index]

条件重复
条件重复于一个行为是被提供由一个消息到一个块带着选择器 whileTrue：和另一个块作为参数。
接受者块发送给它自己那消息 value 且，如果响应的时 true，它发送 value 给另一个块这时重新，发送给它自己 value 又一次。
当那接受者的响应到 value 变成 false，它停止重复且从 whileTrue：消息返回。
例如，条件重复可能被用来去初始化一个名为 list 的数组的所有元素。
index：= 1.
[index<= list size]
      whileTrue: [list at: index put:0.
                  index := index + 1]
块也理解一个消息带着选择器 whileFalse：其重复执行参数块只要接受者的值是 false。
所以，如下表达式是根上面的等价。
index := 1.
[index > list size] 
      whileFalse: [list at: index put: 0.
                   index := index + 1]

程序员是自由的去选择任意消息使得那重复意图最明显。
whileTrue：和 whileFalse 的返回值永远都是 nil。

块参数
为了使一些无序列的控制结构容易表达，块可能带有一个或者多个参数。
块参数被指定通过包括标识符前缀一个冒号在块开始的地方。
块参数被分开与表达式组成那块由一个竖杠。
如下两个例子描述了块带着一个参数。
[:arry | total := total + array size]
和
[ :newElement |
  index := index + 1.
  list at: index put: newElement]
一个常见的使用于带参数的块是去实施功能其被应用到所有元素于一个数据结构。
例如，

p36
许多对象代表不同种类的数据结构响应到消息do：，其带着一个单参数块作为他的参数。
对象接收一个 do：消息求值那块一次为每一个元素包含在达数据结构中的。
每一个元素都被作为块参数而进行一次块求值。
如下的例子计算四方形的值于前5个。
结构是 sum 得值。
sum := 0.
#(2 3 5 7 11) do: [ :prime | sum := sum + (prime * prime)]
 消息 collect：创建一个集合于那值被产生由那块当由接收者的元素提供。
如下表达式的值是一个前5个的四方形的数组。
#（2 3 5 7 11） collect：[：prime | prime * prime]
对象实施这些控制结构提供那值于块参数通过发送给块 value：消息。。
一个块带着一个块参数响应到 value：通过设置那块参数到那 value:的参数且然后执行块内的表达式。
例如，求值如下的表达式到时变量 total 的值为7.
sizeAdder := [:array | total := total + array size].
total := 0.
sizeAdder value: #(a b c).
sizeAdder value: #(1 2).
sizeAdder value: #(e f).
块可以大有多于1个参数。例如
[:x :y | (x * x) + (y * y)]
或者
[:frame :clippingBox | frame intersect: clippingBox]
一个块必须有同样地数目的块参数同那 value：关键字在求值它的消息里。 
上面两个块将被求值通过一个双关键字的消息带着选择器 value:value：。
那两个参数于娜消息指定那两个块参数的值按顺序。
如果一个块接收到一个求值消息带着一个不同的数目跟块参数所带参数数目，一个错误将会被报告。

p37
术语总结
表达式语法被总结在背面本书。
表达式：描述对象的一系列字符。
字面量：描述常量的表达式，如一个数或者一个字符串。
符号：一个字符串其字符序列被保证与其它任何符号都不一样。
数组：一个数据结构其元素被绑定与整数索引。
变量名：一个表达式描述一个变量的当前值。
赋值：一个表达式描述一个变量值的改变。
伪变量名：一个表达式类似于一个变量名。但是，不同于变量名，伪变量名的值不能够被赋值语句改变。
接收者：一个消息被送往的对象。
消息选择器：一个消息要求他的接收者所执行的操作类型的名称。
消息参数：为一个操作指明附加信息的对象。
一元参数：没有参数的消息。
关键字：带着一个分号的标识符。
关键字消息：有一个或多个由一个或多个关键字所组成的参数的消息。
二元消息：一个消息带着一个参数其选择器被一个或两个特殊字符组成。
层叠：一个描述于若干消息到一个对象在一个单一表达式中。
块：一系列被延迟的行为的表述。
块参数：一个参数其必须被提供当特定块被求值。
value 一个消息到一个块要求他执行一套行为他所代表的。
value： 一个关键字被使用在一个消息里去到一个块其块带有块参数；对应的消息要求那块去执行他的一套行为。
ifTrue：ifFalse：消息到一个布林要求条件选择。
ifFalse：ifTrue：消息到一个布林要求条件选择。
ifTrue：消息到一个布林要求条件选择。
ifFalse：消息到一个布林要求条件选择。
whileTrue：消息到一个布林要求条件重复。
whileFalse：消息到一个布林要求条件重复。
do:一个消息去到一个集合要求历遍他的元素。
collect：消息到集合要求变化它的元素。

第三章 类和实例
协议描述
 消息门类
实施描述
变量声明
 实例变量
 共享变量
方法
 参数名
 返回值
 伪变量 self 
 临时变量
原始方法
术语总结

p40
对象代表那组件于微言-80系统--数字，数据结构，进程，磁盘文件，进程计划，文本编辑器，编译器，和应用。
消息代表交互之间于微言-80的组件们--数学计算，数据访问，控制结构，文件操作，文本编辑，编译，和使用应用。
消息使得一个对象的功能对于其他对象可访，同时保持对象的实施被隐藏。
前面的章节介绍一个表达式语法为了描述对象和消息，专注在如何消息被使用去访问一个对象的功能。
这个章节介绍描述方法和类的语法为了展示一个对象的功能是如何被实施的。

每一个微言-80对象是一个实例于一个类。
一个类的实例都有相同的消息界面；类描述了如何去执行每一个操作其可访问的通过界面。每一个操作被描述由一个方法。
选择器于一个消息决定了什么类型的操作那接收者可以执行，所以一个类有一个方法对应每一个在他界面的选择器。
当一个消息被发送给一个对象，那方法绑定与那种类型的消息在接收者的类里地被执行。
一个类也描述什么类型的私有记忆他的实例将有。

每一个类有一个名字其描述他的实例代表的组件的类型。
类名服务于两个基础目的；他是一个简单的方式给实例们去辨认他们自己，和他提供一个方法去指向类在表达式中。
因为类是组件于微言-80系统，他们都被对象代表。
一个类的名称自动变成一个全局分享的变量名。
那个变量的值就是代表这个类的对象。
因为类名是共享变量名，他们必须被大写开头。

新对象被创生通过发送消息到类。
大部分类通过创生一个他们自己的新实例来响应一元消息 new。
例如
 OrderedCollection new
 返回一个新的集合其是系统类 OrderedCollection 的一个实例。
新 OrderedCollection 是空的。
一些类创生实例来响应其他消息。
例如，类其实例代表一天中时间的时 Time；Time 响应到消息 now 通过一个实例代表当前时间。
类其实例代表日期在一年中是 Date；Date 响应到消息 today 通过一个实例代表当下日期。
当一个新实例被创生，它自动分享那方法于那类其接收实例创生消息的。

p41
这个章节介绍两种方式去展示一个类，一种是描述实例的功能另一种是描述那些功能的实施。

1.一个协议描述列出了消息在实例的消息界面。
每一个消息伴同这一个标注描述那操作一个实例将会执行的当他接收到那种类型的消息。

2.一个实施描述展示如何那功能被描述的在那协议描述被实施。
一个实施描述提供那形式于那实例的私有记忆和那一套方法其描述实例如何执行他们的操作。

第三种方式去展示类在一个互动视角叫做一个系统浏览器。
那浏览器是那编程界面的一部分且被用在一个运行着的微言-80系统。
协议描述和实施描述被设计成为了非互动文档如这本书。
那浏览器将被描述简要的在第17章。

协议描述
一个协议描述列出了那消息可以被一个特定类的实例所理解的。
每一个消息被列出带着一个注释关于他的功能。
注释描述了那操作将会被执行的当那消息被收到和什么值会被返回。
注释描述什么会发生，而不是怎样那操作会被执行。
如果注释没有给出决定于被返回的值，那么返回值就被假定为消息的接收者。

例如，一个协议描述条目为了那消息到一个 FinancialHistory 带着选择器 spend： for：是
spend: amount for: reason
记住一个钱的数目 amount，已经被花费为了 reason。
消息在一个协议描述被描述以消息样式的形式。
一个消息样式包含一个消息选择器和一套变量名，一个名字对应一个参数其带着那个选择器的消息会有的。
例如，消息样式
spend: amount for: reason

 p42
	
匹配消息被描述由如下三个表达式。
HouseholdFinances spend: 32.50 for: 'utilities'
HouseholdFinances spend: cost + tax for: 'food'
HouseholdFinances spend: 100 for: usualReason
参数名被使用在注释去指向那参数。
上述例子中的注释决定了第一个参数代表被花费的钱数和第二个参数代表钱被花在了什么上。

消息门类
消息其召唤类似操作的被分组一起在一个门类。
门类有名字其指示共同功能于消息在那一组。
例如，发送到 FinancialHistory 的消息被分为3个门类名为 transaction recording，inquiries，和 initialization。
这个分门别类意在使得协议更可读对于用户；他不会影响类的操作。

完整地协议描述为了 FinancialHistory 被展示如下。

FinancialHistory protocol

transaction recording
  receive: amount from: source
记忆一笔钱 amount，被收到从 source。
  spend: amount for: reason
记忆一笔钱 amount，被花费为 reason。

inquiries
  cashOnHand
返回所有手头现金。
  totalReceivedFrom: source
返回目前为止所有从 source 收到的钱数。
  totalSpentFor：reason
 返回目前为止总数的花费为了 reason 。

一个协议描述提供足够的信息给一个程序员去知道如何使用那个类的实例。
从如上协议描述，我们知道任何 FinancialHistory 的实例应该相应于消息其选择器是receive:from:, spend:for:, cashOnHand, totalReceivedFrom:, totalSpentFor:, 和initialBalance:.
我们可以想象当我们第一次创建一个实例于 FinancialHistory，消息 initialBalance： 应该被发送到那实例为了设置他的变量值。

p43

实施描述
一个实施描述包括三个部分。
1.一个类名。
2.一个变量声明可到达实例。
3.实例用来相应消息的方法。

一个例子于一个完成的实施描述为了 FinancialHistory 被给出如下。
在一个实施描述中的方法被分成同样地门类在协议描述中。
在互动系统浏览器，门类被用来提供一个层级查询路径为了访问一个类描述的部分。
没有特别字符界定分开那不同部分于实施描述。
字体和加重表明不同的部分。
在互动系统浏览器，部分被储存独立的且系统浏览器提供一个结构的编辑器为了访问他们。

class name FinancialHistory
instance variable names cashOnHand
                        incomes
                        expenditures
instance methods
transaction recording
 receive:amount from:source
  incomes at:source
          put:(self totalReceivedFrom:source)+amount
  cashOnHand:=cashOnHand+amount

 spend:amount for:reason
  expenditures at:reason
               put:(self totalSpentFor:reason)+amount
  cashOnHand:=cashOnHand-amount

inquiries
 cashOnHand
  ^cashOnHand

 totalReceivedFrom:source
  (incomes includesKey:source)
    ifTrue:[^incomes at: source]
    ifFalse:[^0]


initialization

initialBalance:amount
 initialBalance:amount
  cashOnHand:=amount.
  incomes:=Dictionary new.
  expenditures:=Dictionary new

这个实施描述不同于那个展示 FinancialHistory 在本书内首封面的。那个在内首封面的有一个添加的部分标为“类方法”将会被解释在第五章；同样，他忽略了初始化方法被展示在这里的。

变量声明
一个类中的方法有路径到5个不同类型的变量。
这些类型的变量不同在多广泛他们可以到达（他们的域）和多长久他们存在。

有两种私有变量只对一个对象可访问。
1.实例变量存在在那个对象的整个生命周期。
2.临时变量被创生为了一个指定的行为且可方位只对那个行为的时间内。

实例变量代表一个对象的当下状态。
临时变量代表那暂时状态其必须得去实行某些行为。
临时变量典型的被绑定与一个单个方法的执行：他们被创生当一个消息导致那方法被执行且被丢弃当那方法完成通过返回一个值。

有三中其他种类的变量可以被访问由一个以上的对象。
他们被区分出来通过有多广泛他们被分享。
3.类变量被共享由一个类的所有实例。
4.全局变量被共享由所有类的所有实例（即是，所有对象）。
5.池变量被共享由系统中一个类的子集的实例。

大多数共享变量在那系统中不是类变量就是全局变量。
大部分全局变量指向系统类。


p45

一个FinancialHistory 的实例名为HouseholdFinances 被使用在若干例子中在前几章。
我们使用 HouseholdFinances 就像他被定义为一个全局变量名。
全局变量被用来指向对象其不是另外对象的一部分。

回忆共享变量3-5是开头大写的，同时私有变量1-2不是。
一个共享变量的值是独立的于实例正在使用那方法其出现那个变量名。
实例变量和临时变量的值依靠使用那个方法的实例，就是，接收到一个消息的实例。

实例变量
有两种实例变量，命名的和索引的。
他们不同在他们是如何被声明和如何被访问的。
一个类可能只有命名的实例变量，只有索引的变量，或都有一些。

命名的实例变量
一个实施描述包括一套实例变量名其组成了那单独的实例。
每一个实例由一个变量对应每一个实例变量名。
在 FinancialHistory 的实施描述中变量声明指定了3个实例变量名。
instance varible names cashOnHand
                       incomes 
                       expenditure
一个 FinancialHistory 的实例使用两个字典去保存全部数额花费和收入为了多种原因，并且用另一个变量去跟踪手上现金。
*支出指向一个字典其绑定花费原因与花费数额。
*收入指向一个字典其绑定收入来源与收入数额。
*手上现金指向一个数目代表可用钱数。

当类方法的表达式使用变量名 incomes，expenditures，或者 cashOnHand之一时，这些表达式指向接收到消息的实例的对应实例变量。

当一个新的实例被创建通过发送一个消息到类，他会由一套新的实例变量。
那实例变量被初始化根据绑定与实例创建消息的方法所指定的。
默认初始化方法赋予每一个实例变量一个 nil 值。

p46

例如，为了那前一个例子消息到 HouseholdFinances 正常工作，一个表达式如下面的必须已经被求值。
HouseholdFinances:=FinancialHistory new initialBalance:350
FinancialHistory new创建一个新的对象其三个实例变量都指向 nil。
那initialBalance:消息到那新实例给那三个实例变量更合适的初始值。

索引的实例变量
一些类的实例可以有实例变量其不被访问通过名字。
这些被叫做索引的实例变量。
取代被指向通过名字，索引的实例变量被指向通过消息其包括整数，被叫做索引，作为参数。
因为索引是一个形式于绑定，那两个基础的索引消息由一样的选择器如同那绑定消息到字典--at：和 at：put：。
例如，Array 的实例有索引的变量。
如果 names 是 Array 的一个实例，表达式
 names at：1
返回他的第一个索引的实例变量值。表达式
 names at:4 put:'Adele'
 保存那字符串‘Adele’作为那值于第四个索引的实例变量于 names。
合法的索引从1到实例中索引到的数目。 

 如果一个类的实例有索引的实例变量，他的变量声明将包括一行indexed instance variable。
例如，系统类 Array 的一部分实施描述是
class name Array
indexed instance variables

每一个实例于一个类其允许索引的变量可能有不同的数目于他们。
FinancialHistory 的所有实例有三个实例变量，但是 Array 的实例可能有任何数目的实例变量。

一个类其实例有索引的实例变量也可以有命名的实例变量。
这种类的所有实例将有同样数目的命名的实例变量，但是可能有不同数目的索引的变量。
例如，一个系统类代表一个集合的其元素是有序的，OrderedCollection，有索引的实例变量去保存他的内容。
一个 OrderedCollection 可能有更多空间为了保存元素多余当下被使用的。
 
p47

那两个命名的实例变量记住那第一个和最后一个元素的索引在内容中。
class name OrderedCollection
instance variable names firstIndex
                        lastIndex
indexed instance variable

所有 OrderedCollection 的实例将有两个命名的变量，但是一个可能有五个索引的实例变量，另一个有15个，另一个18，等等。

FinancialHistory 的一个实例的命名的实例变量是私有的意味着到访那变量值的路径被那实例所控制。
一个类可能有也可能没有包括消息给出直接访问路径到那实例变量。
索引的实例变量不是私有的在这种意义上，因为直接访问到那变量的值是可行的通过发送消息带着选择器 at：和 at：put：。
鉴于这些消息是唯一的方式去访问索引的实例变量，他们必须被提供。

带有索引的实例变量的类创建新的实例通过消息 new：取代通常的消息 new。
new：的参数告诉被提供的索引的变量的数目。
list：=Array new：10
创建一个数组带着10个元素，每一个被初始化为特殊对象 nil。
一个实例的索引的变量数目刻意被找到通过发送给他消息 size。
对消息 size 的响应
 list size 
是，在这个例子里，是整数10。

求值如下的每一个表达式，按顺序。
list:=Array new:3.
list at:1 put:'one'.
list at:2 put:'two'.
list at:3 put:'three'
是等价的与那单个表达式
list:=#('one' 'two' 'three')

共享变量
变量其被共享由多于一个对象的成组出现叫做池。
每一个类有两个或更多池其变量可以被他的实例访问。
一个池被所有类共享且容纳全局变量；这个池名为Smalltalk(微言)。

p48

每一个类也有一个池其只可访问对于他的实例并且容纳类变量。
这两个法定池之外，一个类可能访问一些其他特别目的池其被若干类共享。
例如，有若干类在系统中其代表文本信息；这些类需要共享 ASCII(American Standard Code for Information Interchange美国标准信息交换编码)字符代码为了字符其不容易视觉的决定，如一个回车，制表符，或者空格。
这些数目被包括作为变量在一个池名为 TextConstants 其被实施文本显示和文本编辑的类所共享。

如果 FinancialHistory 有一个类变量名为 SalesTaxRate 且共享一个池字典其名为 Fi nancialConstants，那声明将会被表达如下。
instance variable names cashOnHand
                        incomes
                        expenditures
class variable names    SalesTaxRate
shared pools            FinancialConstants
SalesTaxRate 是一个类变量的名称，所以他可以被用在类的任何方法中。
FinancialConstants，在另一方面，是一个池的名字；他是在池中的变量可以被用在方法中。
为了声明一个变量为全局的( 让所有类和用户互动系统都知道），那变量名必须被作为一个键插入到字典 Smalltalk。
例如，要让 AllHistories 全局化，求值表达式
 Smalltalk at:#AllHistories put:nil
这时再用一个赋值语句去设定 AllHistories 的值。

方法
一个方法描述如何一个对象将执行一个他的操作。
一个方法被组成由一个消息样式和一系列表达式被分隔由点号。
下面的例子方法描述了那响应于一个 FinancialHistory 到消息其通知他于支出。
spend:amount for:reason
 expenditures at:reason
              put:(self totalSpentFor:reason)+amount.
 cashOnHand:=cashOnHand-amount
那消息样式，spend:amount for:reason，决定了这个消息将被用来响应所有消息带着选择器spend:for:。

p49
第一个表达式在主体于这个方法加上那新数目到那书目已经花费在了那原因被指定的。
第二个表达式是一个赋值语句其减少那值于手上现金以新数目。

参数名
消息样式被介绍在这一章的早些时候。
一个消息样式包括一个消息选择器和一套参数名，每一个带有那个选择器的消息应有的参数都对应一个。
一个消息样式匹配任何消息其有那同样地选择器。
一个类将只有一个方法带着一个给定的选择器其在他的消息样式。
当一个消息被发送，那方法带着匹配的消息样式被选定从那接收者的类中。
那表达式在那被选定的方法被求值一个接着一个。
在所有表达式被求值之后，一个值被返回给那消息的发送者。

变量名被发现在一个方法的消息样式是伪变量名器指向那实际消息的参数。
如果如上的方法被召唤由表达式
 HouseholdFinances spend:30.45 for:'food'
那伪变量名 amount 将指向数30.45且伪变量名 reason将指向字符串‘food’期间于方法内表达式的求值过程。
如果相同的方法被召唤由表达式
HouseholdFinances spend:cost+tax for:'food'
cost将被送那消息+tax 且那值他返回的将被指向作为 amount 在那方法。
如果 cost指向100且 tax 指向6.5，那值于 amount 将是106.5。

因为参数名是伪变量名，他们可以被用来访问值如变量名，大师他们的值不能被改变通过赋值语句。
在方法为了spend:for:，一个语句于那形式
 amount:=amount*taxRate
将是句法的错误因为 amount 的值不可以被重赋。

返回值
方法为了spend:for:不指定什么值于那消息应该是。
因此，那默认值，那接收者自身，将会被返回。
当另一个值被指定，一个或多个返回表达式将被包括在那方法内。
任何表达式可以被变成一个返回表达式通过前缀他用上箭头^。
一个变量值可能被返回如
^cashOnHand
另一个消息的值可以被返回如
^expenditures at:reason
一个字面量对象可以被返回如
^0
甚至一个赋值语句可以被变成一个返回表达式，如
^initialIndex:=0
赋值被先执行。
变量的新值这时被返回。

一个例子于使用一个返回表达式是如下的实施于totalSpentFor:。
totalSpentFor:reason
 (expenditures includesKey:reason)
   ifTrue:[^expenditures at:reason]
   ifFalse:[^0]
这个方法组成于一个单条件表达式。
如果那支出 reason 在 expendtures，那绑定的值被返回；否则，0被返回。

伪变量 self
一起与伪变量被用来指向那消息的参数，所有方法由路径去一个伪变量名为 self 其指向那消息接收者自身。
例如，在方法匹配spend:for:，那消息totalSpentFpr:被发送到那消息spend:for:的接收者。
spend:amount for:reason
 expenditures at:reason
              put:(self totalSpentFor:reason)+amount.
 cashOnHand:=cashOnHand-amount
当这个方法被执行时，第一件事情发生的时totalSpentFor:被送达同一个对象(self)其接收spend:for:。
这个消息的结果被送消息+amount，并且那消息的结果被用作at:put:的第二个参数。

p51

伪变量 self 可以被用作实施递归功能。
例如，消息factorial被理解由整数为了去计算那合适的功能。
绑定于factorial的方法是
 factorial
  self=0 ifTrue:[^1].
  self<0
    ifTrue:[self error:'factorial invalid']
    ifFalse:[^self*(self-1) factorial]
接收者是一个整数。
第一个表达式测试看是否接收者是0和，如果他是，返回1.
第二个表达式测试那符号于那接收者因为，如果他小于0，程序员应该被通知一个错误（所有对象响应到消息 error:用一个报告说一个错误被遭遇到了）。
如果接收者大于0，这时那被返回的就是
self*(self-1) factorial
被返回的结果是接收者被乘以比接收者小一个数的乘阶。

临时变量
参数名和 self 是可访的只有期间于一个方法的单独执行。
另外 于这些伪变量名，一个方法可能要求一些其他变量为了在执行期间使用。
这些被叫做临时变量。
临时变量被指定通过包括一个临时变量声明之间于消息样式和那方法的表达式。
一个历史声明被组成于变量名之间于竖杠。
方法为了spend:for:可以被重写去使用一个临时变脸去保存那前一个支出。
spend:amount for:reason
 | previousExpenditures |
 previousExpenditures:=self totalSpentFor:reason.
 expenditures at:reason
              put:previousExpenditures+amount.
 cashOnHand:=cashOnHand-amount.
临时变量值可访问只对于方法中的语句并且当方法完成执行就被忘记。
所有临时变量被初始化的指向 nil。

在互动于微言-80系统，程序员可以测试算法其使用临时变量。
测试可以被执行通过使用竖杠标记去声明变量为了仅仅在立即求值的期间。

p52

假设表达式去尝试的包括参考到变量 list。
如果变量 list 没有被声明，一个尝试去求值那表达式将创造一个语法错误消息。
替代的，程序员可以声明 list 为一个临时变量通过前缀那表达式带着那声明|list|。
那表达式被分隔由点号，正如在一个方法中的语法。
| list|
list:=Array new:3.
list at:1 put:'one'.
list at:2 put:'four'.
list printString.
程序员互动的选择所有五行--那声明和那表达式--并要求求值。
变量 list 可访问只期间于娜单次执行于那被选择的部分。

原始方法
当一个对象接收到一个消息，他典型的只是发送其他消息，所有哪里是一些事情真正发生的呢？
一个对象可能改变他的实例变量的值当他收到一个消息，其当然符合“一些事情发生”。
但是这个很难被看做足够。
事实上，他不足够。
所有行为在那系统中被召唤由消息，然而，所有消息不被响应由执行微言-80方法。
有大约100个原始方法其微言-80虚拟机知道如何去执行。
例子于消息其召唤原始方法的时那+消息到小整数，at：消息到带有索引的实例变量的对象，和new 和new:消息到类。
当3得到那消息+4，他不执行一个微言-80方法。
一个原始方法返回7作为那个消息的值。
完全的套于原始方法被包括在第四部分于本书，其描述那虚拟机。

方法被执行作为原始方法开始于一个表达式于形式
<primitive #>
哪里#是一个整数指定哪一个原始方法将被接着。
如果那原始方法无法正确执行，执行继续在那微言-80方法。
那表达式<primitive #>被跟随由微言-80表达式其处理失败情况。

p53
术语总结
 class 类 一个对象其描述那实施于一套相似的对象。

instance 实例 一个对象之一其被描述于一个类；他有记忆且能对消息响应。

instance variable 实例变量 一个变量可访问对于一个单一对象在那对象的整个生命周期；实例变量可以被命名或则索引化。

protocal description 协议描述 一个描述于一个类为了他的实例的公共消息协议。

implementation description 实施描述 一个描述于一个类为了他的实例私有记忆和安一套方法其描述如何实例们这行他们的操作。

message pattern 消息样式 一个消息选择器和一套参数名，带着这个选择器的消息必须有的参数都有一个。

temporary variable 临时变量 一个变量被创建为了一个特别的行为且可以访问只在那个行为期间内。

class variable 类变量 一个变量共享由所有实例于一个单独类。

global variable 全局变量 一个变量共享于所有类的所有实例。

pool variable 池变量 一个变量共享于一套类们的实例们。

Smalltalk 微言池 一个池被共享由所有类们保存有那全局变量。

method 方法 一个过程描述如何去执行一个对象的一个操作；他被组成由一个消息样式，临时变量声明，和一系列表达式。
一个方法被执行当一个消息匹配他的消息样式被发送到一个类的实例于那类这个方法被找到。

argument name 变量名 一个伪变量名可以到一个方法只为了那方法的执行期间；那参数名的的值是那召唤这方法的消息的参数。

^ 返回值 当被用在一个方法，决定了下一个表达式的值将是拿方法的值。

self 自身 一个伪变量名指向一个消息的接收者。

message category 消息门类 一组方法在一个类描述中。

primitive method 原始方法 一个操作直接由微言-80虚拟机执行的；他不被描述像一系列微言-80表达式。

第四章
子类

子类描述
一个例子子类
方法判定
 消息到自身
 消息到上级
抽象上级类
子类框架消息
术语总结

p56

微言-80系统中的每一个对象是一个类一个实例。
一个类的所有实例代表同一种类的系统组件。全例于一类表同型统件。类之诸例乃同型统件。
举，诸例于 Rectangle 表一矩形且诸例于 Dictionary 表一套联间于名值。
其实例于一类皆表同型件现于彼径例应于讯及于彼形诸例之变。
*诸例于一类应于同套讯且使同套法作此。
*诸例于一类拥同数名例变且使同名指彼。
*一象可拥索例变仅若诸例于其类可拥索例变。

类构如述至今无详供为任交于类员。
每象乃一例于一确类。
此构被绘于图4.1.
于彼图，诸圈表例及箱表类。
若圈内于箱，则其表一例于类，其类表由彼箱。

图4.1

憾于交于类员乃一限于计于一象面统因其勿允任共间于类绘。
吾设欲两例为甚仿，然异于某面。
举，一浮点数与一整数仿于其机于应于算讯，然异于彼路其表值。
一序集与一袋仿于其乃筐于其员可被添且于其员可被删，

p57

然其异于细路其独员被访。
异间彼似象可为外显，如应于许异讯，或纯内，如应于许同讯由行异法。
若类员勿允叠，此型分似间于两象不可被保由彼统。

彼最凡路为克此限乃允强交于类线。（图4.2）

吾谓此法多承。
多承允况于其许象乃例属两类，时余象乃例属仅一类或其余。
一稍凡弱于彼无交限于类乃允一类涵诸例属另类，然勿允更杂分。（图4.3）
吾谓此法子类。
此循语程言 Simula，其涵一似念。
子类严承；若任例于一类乃例属另类，则诸例于此类必为例属彼另类。

 p58

微言-80统供子类型承为其类。
此章绘何子类改其父类，何此撼彼联于讯及法，及何子类术供一框为类于统。

子类绘
一子类述其诸例将同于诸例属一另类，谓其父类，除其异乃被确陈。
微言-80统恒生一新类乃一子类于一旧类。
一统类名象绘彼似于诸象属彼统，故诸类终为一子类于象。
一类绘（协或施）确何其例异于彼例属其父类。
诸例于一超类不被篡由子类之展。

一子类乃是具格类且故可有子类于自。
诸类有一父类，因多类或分彼同父类，故彼诸类成一树架。
一类拥串类由彼其承变与法。
此串始于其父类且续于其父类之父类，类推。
彼承链续穿父类系至逢象。
象乃独根类；其乃独类其无父类。

忆其一施绘拥三基部：
1.一类名
2.一变申
3.一套法
一子类须供一新类名为己，然其承变申与法属其父类。
新变或被申且新法或被添由彼子类。
若例变名被添在子类变申，子类例将有例变多于父类例。
若分变被添，其将被访由子类诸例，但非父类诸例。
诸变名被添须异于任被申于彼父类。

p59

若一类无由索例变，一子类可申其诸例将有索变；此索变将被添到任被承名例变。
若一类有索例变，其子类须亦有索例变；一子类可亦申新名例变。

若一子类添一法其讯式有同择如一法于彼父类，其诸例将应于讯携彼择以行彼新法。
此谓覆盖一法。
若一子类添一法携一择未见于父类之法，子类诸例将应其讯彼未被懂于父类诸例。

综，每部于一施绘可被改由一子类于一异路：
1.类名须被覆。
2.变或被添。
3.法或被添及覆。

一子类样
一施绘涵一录，未现于前章，定其父类。
下样乃一类被生为一子类属 类 FinancialHistory被介于章3.
诸例于子类分机于 FinancalHistroy 为储信于币费及入。
彼有添机于储迹于费其乃扣税。
子类供强新类名（DecuctibleHistory），且添一例变及四法。
一之此法（initialBalance：）覆一法于父类。

彼类绘为 DeductibleHistory 随。
class name DeductibleHistory
superclass FinancialHistory
instance variable names deductibleExpenditures
instance methods

transaction recording
 spendDeductible:amount for:reason
  self spend:amount for:reason.
  deductibleExpenditures:=
   deductibleExpenditures+amound
 spend:amount for:reason deducting:deductibleAmount
  self spend:amount for:reason.
  deductibleExpenditures:=
   decuctibleExpenditures+deductibleAmount

inqurires
 totalDeduction
  ^deductibleExpenditures

initialization
 initialBalance:amount
  super initialBalance:amount
  deductibleExpenditures:=0


为知诸讯懂由一例于 DeductibleHistory，其必测协于 DecuctibleHistory，FinancialHistory，及 Object。
诸例于 DeductibleHistory 拥四变--三承于父类 FinancialHistory，一定于类 DecuctibleHistory。
类 Object 申无例变。

图4.4
定 DeductibleHistory 乃一子类于 FinancialHistory。
每箱于此图被标于左上角用其表类之名。

诸例于 DeductibleHistory 可被用于录迹于物其缴税（人，房，商）。
诸例于 FinancialHistory 可被用于录迹于物勿缴税（善会，教会）。

p61

实，一例于 DeductibleHistory 可被用于替一 FinancialHistory 例勿测因其应于诸同讯以同式。
添到讯及法承于 FinancialHistory，DeductibleHistory 之一例可应于讯定彼诸或部于一费乃课税。
彼新讯具乃spendDeductible:for:，其被用若总数乃税；spend:for:deduction:被用若仅部于费乃税。
总税收可被查于送 DeductibleHistory 彼讯 totalDeductions。

择法
当一讯被送，彼法于收类寻一法配择。
若无有，彼诸法于彼类之父类续被查。
彼寻续于父类链至一配法得见。
设吾送一例于 DeductibleHistory 以讯携择 cashOnHand。
彼寻为适法以行始于收类，DeductibleHistory。
当其未获，彼寻续于视 DeductibleHistory 之父类，FinancialHistory。
当一法携择 cashOnHand 被获，彼法被行作应到彼讯。
彼应到此讯乃返例变 cashOnHand 之值。
此值被获于彼讯之收，即，于彼例于 DeductibleHistory。

彼寻为一配法循父类链，终于类 Object。
若无配法寻获于任类于父类链，彼收被送讯doesNotUnderstand:,彼参乃彼棘讯。
有一法为择doesNotUnderstand: 于象其呈错于人。

设吾送一例于 DeductibleHistory 一讯携择spend:for:。
此法被获于父类 FinancialHistory。
彼法，示于章3，为

spend:amount for:reason
 expenditures at:reason
              put:(self totalSpentFor:reason)+amount.
 cashOnHand:=cashOnHand-amount.
彼值于例变（expenditures 和 cashOnHand）被获于讯收，彼例于 DeductibleHistoru。

p62

彼伪变 self 亦被指于此法；self 表彼 DeductibleHistory 例其乃彼讯收。

讯之我
当一法涵一讯其收乃我，彼寻法为讯始于彼例类，略谁类涵彼法具我。
故，当信self totalSpentFor:被估于法配spend:for:获于 FinancialHistory，彼寻为彼讯绑定讯择totalSpentFor:始于 self 之类，即，在 DeductibleHistory。

讯到 self 将被释由两样类名One和Two。Two 乃一子类于 One 且 One 乃一子类于 Object。
两类涵一法为讯test。
类 One 亦涵一法为讯 result1其返信 self test 之果。
class name One
superclass Object
instance methods
 test
  ^1
 result1
  ^self test

class name Two
superclass One
instance methods
 test
  ^2
一例于每类将被用于验彼法决为讯 self。
样1乃一例于类 One 且样2乃一例于类Two。
example1:=One new.
example2:=Two new.
彼联间 One和 Two 被示于图4.5。
添于标彼箱为定类名，许圈亦被标为定一名指其例。

p63

下表示果于变信估。
expression                result
example1 test             1
example1 result1          1
example2 test             2
example2 result1          2

彼两result1讯唤彼同法，其被获于类 One。
彼生异果因讯至 self 涵于彼法。
当 result1被送至 example2，彼寻为一配法始于 Two。
一法未获于 Two，故彼寻续由视彼父类，One。
一法为 result1得获于 One，其构由一信，^self test。
彼伪变 self 指 example2之收。彼寻为应于 test，故，始于类 Two。
一法为 test 被获于 Two，彼返2。

信至父
忝伪变名 super可访为用于一法信。
彼伪变super指讯收，一如self行。
然，当一讯送至super，彼寻为一法非始于彼收类。
替，彼寻始于父类于彼类涵此法。
使super允一法访诸法定于一父类勿论彼法已被覆于子类。

p64

 用 super 为非收（样，为一参），无异于用 self；用 super 仅撼始类于其讯被寻。

 讯到super将被释用两多样类名 Three 和 Four。
Four 乃一子类于 Three，Three 乃一子类于前样 Two。
Four 覆法应讯test。
Three涵法应两新讯--result2返信self result1之果，和result3返信super test之果。
class name Three
superclass Two
instance methods
 result2
  ^self result1
 result3
  ^super test

class name Four
superclass Three
 test
  ^4

诸例于One,Two,Three和Four均可应于讯test和result1.
彼应于诸例属Three和Four到讯绘彼效于super(图4.6）。
example3:=Three new.
example4:=Four new.
一试去发讯result2或result3到example1或example2乃一错因诸例于One或Two不懂讯result2或result3.

下表示彼果于送诸讯。
expression           result
example3 test        2
example4 result1     4
example3 result2     2
example4 result2     4
example3 result3     2
example4 result3     2

p65

 当test被送到example3,彼法于Two被用，因Three未覆彼法。
example4应result1以4同因与example2应以2.
当result2被送到example3,彼寻为一配法始于Thress.
彼法获返信果self result1.
彼寻为应result1亦始于类Three.
一配法未获于Three或其父类，Two。
彼法为result1被获于One且返果self test.
彼讯为应test再始于类Three。
今，彼配法得获于Three父类Two.

彼效于送讯至super将被绘由彼应example3和example4于彼讯result3。
当result3被送至example3,彼寻为一配法始于Three.
彼获法返值属信super test.
因test被送至super，彼寻为一配法非始于类Three,然于其父类，Two.
彼法配test于Two返一2.
当result3被送至example4,其果仍为2，虽Four覆其法配test.

此样亮一潜惑：super未意始寻于收之父类，其，于末样，应乃类Three.
彼意始寻于父类于彼类涵法于其super被用，其， 于末样，乃类Two.
 设Three已覆彼法配test由返3，彼果于example4 result3仍为2.
间，然，父类于彼类其法涵super被获同于彼收之父类。

p66

另样于使super乃于法配initialBanlance:于典。

initialBalance:amount
  super initialBalance:amount.
  deductibleExpenditures:=0

此法覆一法于彼父类FinancialHistory.
彼法于DeductibleHistory构于两新。
首信递控至父以行彼初于款。

super initialBalance:amount

伪变super指收于彼讯,然定寻为法应略DeductibleHistory且始于FinancialHistory。
于此路，信从FinancialHistory勿必复于DeductibleHistory.
彼次信于法做彼子类殊初。

deductibleExpenditures:=0

若self替彼super于initialBalance:法，其将致一无极，因每次initialBalance:被送，其将被再送。

理父类
理父类被创当两类分一部于其绘且无一乃正子类于彼。
一应父类被生为彼两类其涵此分面。
此种父类被谓理因其非创为拥诸例。
以图示前，一类父类表彼情绘于图4.7。
知彼理类非直涵诸例。

如一样于使一理父类，虑2类其诸例表典。
一类，名SmallDictionary,低空被需储其容；另，名FastDictionary,简储名值且使hash术
去找名。
两类使两并表其涵名及联值。


p67

SmallDictionary储名值续且使一简线寻去找名。
FastDictionary简储名值且使一hashing术去找名。
除异于何名被找，此两类极仿：彼分同协且彼同使并表储其容。
此仿被表于一理父名DualListDictionary.
彼系间此三类被示如图4.8.

彼施绘为理类，DualListDictionary被示下。

p68

class name DualListDictionary
superclass Object
instance variable names names
                        values
instance methods
accessing
 at:name
  |index|
  index:=sefl indexOf:name.
  index=0
    ifTrue:[self error:'Name not found']
    ifFalse:[^values at:index].

 at:name put:value
  |index|
  index:=self indexOf:name.
  index=0
    ifTrue:[index:=self newIndexOf:name].
  ^values at:index put:value.

testing
 includes:name
   ^(self indexOf:name)~=0.
 isEmpty
   ^self size=0.

initialization
 initialize
   names:=Array new:0.
   values:=Array new:0.
此绘于DualListDictionary只使讯定于DualListDictionary其本或已绘于此及前章。
彼外协为一DualListDictionary构由讯at:,at:put:,includes:,isEmpty,和initialize.
一新DualListDidtionary（实一新例于一子类于DualListDictionary)被生由送他讯new.
即送讯initialize故彼参可被加入彼两例变。
彼两变乃初空组（Array new:0).

三讯到self使于彼法非施于DualListDictianary--size,indexOf:,和newIndexOf:.
此乃故彼DualListDictionary被谓理。
若一例被生，他将无能胜应诸须讯。
彼两子类，SmallDictionary和FastDictionary,必施彼三逸法。
实彼寻恒始于self指之例类谓父类之法可被定于至self之讯，然彼应法被获于子类。
如此，一父类可供一框为一法其被改或施于子类。

SmallDictionary乃一子类于DualListDictionary其使一低数空去表彼联，然或用长时去寻一联。
他供法为彼三讯其未施于DualListDictionary--size,indexOf:,和newIndexOf:.
其未添变。
class name SmallDictionary
superclass DualListDictionary
instance methods

accessing
  size
    ^names size

private
  indexOf:name
    ^to:names size do:
      [:index|(names at:index)=name ifTrue:[^index]].

  newIndexOf:name
    self grow.
    names at:names size put:name.
    ^names size

  grow
    |oldNames oldValues|
    oldNames:=names.
    oldValues:=values. 
    names:=Array new:names size+1.
    values:=Array new:values size+1.
    names replaceFrom:1 to:oldNames size with:oldNames.
    values replaceFrom:1 to:oldValues size with:oldValues.

因名被储续，彼长于SmallDictionary乃彼长于其名组names。
彼索于一殊名被定有一线寻于彼组名。
若无配得获，彼index乃0，败标于寻。
任一新联被添于典，彼法配newIndexOf:被用去寻适索。
其设彼长于names和values乃确长其需储彼时员。
此意无空可用为添新员。
彼讯grow生两新组其乃复于前组，携一多员于尾。

p70

于法配newIndexOf:,首彼长于names及values被增且彼新名被储于新空位（未员）。
彼法其唤newIndexOf:负责储value.

吾可估彼下样信。

expression                           result
--------------------------------------------
ages:=SmallDictionary new            a new, uninitialized instance
ages initialize                      instance variabls initialized
ages isEmpty                         true
ages at:'Brett' put:3                3
ages at:'Dave' put:30                30
ages includes:'Sam'                  false
ages includes:'Brett'                true
ages size                            2
ages at:'Dave'                       30

每上样信，吾标于何类彼信被获和于何类任讯送至self被获。

message selector     message to self        class of method
-----------------------------------------------------------
initialize                                  DualListDictionary
at:put:                                     DualListDictionary
                     indexOf:               SmallDictionary
                     newIndexOf:            SmallDictionary
includes:                                   DualListDictionary
                     indexOf:               SmallDictionary
size                                        SmallDictionary
at:                                         DualListDictionary
                     indexOf:               SmallDictionary
                     error:                 Object

FastDictionary乃另一子于DualListDictionary.
彼使一hashing术找名。
Hashing需多空，然少时较一线寻。
		   ifFalse:[index:=index\\names size+1]].
诸象应讯hash以一数返。
诸数应讯\\以余值于参返。

p71

class name FastDictionary
superclass DualListDictionary
instance methos

accessing
  size
    |size|
    size:=0.
    names do:[:name|name notNil ifTrue:[size:=size+1]].
    ^size.
    
initialization
  initialize
    names:=Array new:4.
    values:=Array new:4.

private
  indexOf:name
    |index|
    index:=name hash\\names size+1.
    [(names at:index)=name]
      whileFalse:[(names at:index)isNil
                   ifTrue:[^0]
		   ifFalse:[index:=index\\names size+1]].
    ^index.
  newIndexOf:name
    |index|
    names size-self size<=(names size/4)
      ifTrue:[self grow].
    index:=name hash\\names size+1.
    [(names at:index) isNil]
      whileFalse:[index:=index\\names size+1].
    names at:index put:name.
    ^index.
  grow
    |oldNames oldValues|
    oldNames:=names.
    oldValues:=values.
    names:=Array new:names size*2.
    values:=Array new:values size*2.
    1 to:oldNames size do:
      [:index|
        (oldNames at:index) isNil
	    ifFalse:[self at:(oldNames at:index)
	                  put:(oldValues at:index)]].
p72

FastDictionary覆DualListDictionary之施于initialize为行生组其已有许空派（Array new:4).
彼长于一FastDictionary非简长于其某变因Array恒有空员。
故彼长被定由查每员于Array且算彼数非nil。

彼施于newIndexOf:随基计如被用为SmallDictionary除当彼长于一Array被改（双于此样在彼法配grow),每员被确复从彼旧 Arrays进新组故其诸员被重hash.
彼长非恒变如必于SmallDictionary.
彼长于FastDictionary被改只当彼数于空址在names低于一小。
彼小等与25%于彼诸员。

names size - self size<=(names size/4).

子类框讯
一流于程式，一法应勿涵讯至self若彼寻非乃施于类及承于父。
于绘于DualListDictionary,三此讯存--size, indexOf:，和newIndexOf:.
如吾将见于后章，彼能去应size乃承于Object;彼应乃彼数于索例变。
一子类于DualListDictionary被设去覆此法以返数于names于彼典。

一殊讯，subclassResponsibiliry,被定于 Object。
其被用于施彼诸讯其无能被正施于一理类。
即，彼施于size和indexOf:和newIndexOf:,由微言-80约，应为

self subclassResponsibility

应至此讯唤下法定于类Object.

subclassResponsibility
  self error:'My subclass should have overridden one of my messages.'

如此，若一法应已被施与一子类于一理类，彼错报乃一标至人于何缮彼难。
另，使此讯，彼人创理类于其诸讯送至self被施，且于其彼施乃一标至任于哪法必被覆于子类。

p73

如约，若人决一讯承于一理父类应实不被施，彼正路去覆彼承法乃
self shoulNotImplement
彼应至此讯乃唤下法定于类Object.
shouldNotImplement
  self error:'This message is not approprite for this object.'

有许主子类团于微言-80统其使意于创一讯框其施须备于子类。
有类绘多种集（见章9和10）。
彼集类被承排以最享间于诸类其绘似集。
彼施讯subclassResponsibility和shoulNotImplement。
另样于使子类乃彼团于线量及数类（见章7和8）。
  
术语结
subclass 子类 一类承变及法于一旧类。
superclass 父类 彼类其变及法被承。
Object 象 彼类其乃根于树构类团。
overriding a method 覆 定一法于子类配同讯如一法于一父类。
super 父 一伪变其指讯收；异于self于哪始彼寻为法。
abstrack class 理类（抽象类）一类其定协，然未能全施（彼协）；如约，此类勿创诸例。
subclassResponsibility 子类责 一讯报错其一子类应已施一父类讯。
shouldNotImplement 应勿施 一讯报错次乃一讯承于父类然确无能于诸子类例。



第五章
例初
样元
元团
类变初
择法结
术语结

p76

因诸微言-80统件被表由象且诸象乃类例，彼类须被表由一类例。
一类其例乃类谓元。
此章绘彼殊性于元。
样绘何元被用以供例生及普类询。

于早版于微言统，有仅一元，名Class.
其配彼类构绘于图5.1.
如用于章4，一箱标一类且一圈标一例于彼涵其类。
若可，彼箱被标以其所表类名。
知有一圈于象标Class应每箱于彼图。

此方困于诸类讯协被限于同因其被定于同地。
特，讯以生新例乃同于诸类且无能做任殊初求入户。
于单元，诸类应讯new或new:以返一例其例变皆指nil.
于多诸类，nil非好实例变，故新例需被初由送另讯。
人必确每次一new或new:被送，另讯被送至新例故其被正初。
样于此种初被示于章4为SmalltDictionary和FinancialHistory.


p77
彼微言-80统除彼限其诸类其有同讯协由使每类为一例属其自元类。
任一新类被生，一新元类被生为其自动。
元类似于彼诸类因其涵诸法被用由其诸例。
元类异于彼诸类因其非其自元类之例。
替，其乃诸例属Metaclass类。
另，元类无类名。
一元类可被访由送其例单讯class.
样，Rectangle之元类可被指由信Rectangle class.

一元类之诸讯贯供生及初于诸例，及初于诸类变

例之初
每类克应讯其求正初新例。
多元类被需因其初讯异间异类。
样，吾已见彼Time生新例以应讯new和Date创新例以应讯today.
Time now.
Date today.

此信无意于Point,彼类其诸例表两维地。
Point生一新例一应讯携择x:y:且两参定彼座。
此讯亦无意于Time或Date.

Point x:100 y:150

类Rectangle懂许讯其生新例。
一讯携择origin:corner:使Points表彼左上和右下角参。

Rectangle
  origin:(Point x:50 y:50)
  corner:(Point x:250 y:250)

一讯携择origin:exten:使为参彼左上角和一Point表宽与高。
彼同矩可被生由下信。

p78

Rectangle
  origin:(Point x:50 y:50j)
  exten:(Point x:200 y:250).

于微言-80统，Class乃一理超类为诸元类。
Class绘彼普性于类。
每元类添特行至其单例。
元类或添新例生讯如Date,Time,Point,和Rectangle如上，或其可改基new和new:讯以行默初。

彼构于类及例于统，如绘至今被示于图5.2.

于此图，吾定类Object,Metaclass,和Class,和metaclasses为每。
每圈于箱标Metaclass表一元类。
每箱于箱标Class表一子类于Class。
有一箱为每圈于箱标Metaclass.
每箱涵一圈表其例；彼例指Object或一子类于Object,然非诸元类。

一样元类
因有一对一应间一类及其元类，其绘被陈共。
一施绘涵一部署“类法”其示诸法被添由元类。
彼协为元类乃恒被获由寻于类法部于彼施绘于其单例。
由此，讯至类（类法）及讯至类例（例法）被同列为部于彼全施绘。

p79

下新版于施绘为FinancialHistory涵类法。

class name FinancialHistory
superclass Object
instance variable names cashOnHand
                        incomes
			expenditures
class methods
instance creation
  initialBalance:amount
    ^super new setInitialBalance:amount
  new
    ^super new setInitialBalance:0

instance methods
transaction recording
  receive:amount from:source
    incomes at:source
            put:(self totalReceivedFrom:source)+amount.
    cashOnHand:=cashOnHand+amount

  spend:amount for:reason
    expenditures at:reason
                 put:(self totalSpentFor:reason)+amount.
    cashOnHand:=cashOnHand-amount

inquiries
  cashOnHand
    ^cashOnHand
  totalReceivedFrom:source
    (incomes includesKey:source)
       ifTrue:[^incomes at:source]
       ifFalse:[^0]
  totalSpentFor:reason
    (expenditures includesKey:reason)
       ifTrue:[^expenditures at:reson]
       ifFalse:[^0].

private
  setInitialBalance:amount
    cashOnHand:=amount.
    incomes:=Dictionary new.
    expenditures:=Dictionary new.

三改被做到彼施绘。
1.一门于类法名例创被添。
彼门涵法配initialBalance:和new.
由贯，彼门例创被用为类法其返新例。

2.彼门于例法名initialization被除。其曾涵一法配initialBalance:.

3.一门于例法名private被添。
彼门涵一法配setInitialBalance:,此法涵彼同信其于彼被除法配initialBalance:.

此样绘何元类创被初例。
彼例创法配initialBalance:和new无有至径去彼例变于彼新例（cashOnHand,incomes,和expenses).
此因彼法非部于彼新例之类，而乃于彼类之类。
比例创法首创非初例送一初讯，setInitialBalance:,至彼新例。
彼法应此信被或于例法部于FinancialHistory之施绘；其可赋恰值至彼例变。
彼初讯非视部于外协于FinancialHistory故其被门为私。
其贯仅送单且仅由类法。

彼旧初讯initialBalance:被除因彼正路以创一FinancialHistory乃用信如
FinancialHistory initialBalance:350
非
FinancialHistory new initialBalance:350
实，次信将生错因例于FinancialHistory不再被绘如应至initialBalance:.
吾本可持彼例法initialBalance:且施彼类法配initialBalance:唤于其，然吾意勿用同择为例及类法以升彼读于施绘。
然，将无歧若同择被用。

p81

元类团
如他类，一元类承自一父类。
最易路去构元类团乃去做一每子类于Class.
此构被示于图5.2.
Class绘彼普性于类。
每元类添殊行到其例。
元类或添新例创讯或其改彼基new和new:讯以施许默初。

当元类被添入微言-80统，进一步于类构被发。
彼元类子类团被迫并于子类团于彼类其乃自例。
故，若DeductibleHistory乃一子类于FinancialHistory,择DeductibelHistory之元类必为一子类于FinancialHistory之元类。
一元类贯由仅一例。

一理类名ClassDescription被供去绘类及其例。
Class和Metalcass乃子类于ClassDescription.
因彼父类链于诸象终于Object且Object无父类，Object之元类的父类乃Class.
从Class，彼元类承讯其供协为生例（图5.3).

彼父类链从Class直到类Object.
知彼承于箱标Object class如彼承于箱标Object;此仿绘彼并承。
一全绘于此部于统，涵彼系间Mataclass及其元类，被供于章16.

一样于元类承团，虑彼施于initialBalance:于FinancialHistory类.

initialBalance:amount
  ^super new setInitialBalance:amount

此法生意新例由估彼信super new;其使彼法配new获于父类类法，非本类类法。其此送彼新例讯setInitialBalance:携彼初值于费为参。
似，new被重施为创一例使super new被随setInitialBalance:.

new
  ^super new setInitialBalance:0.

p83

何乃彼法配讯new至super被获？
彼子类承于远离恶并彼承于其例。
若一类乃一子类于另，其元类将为一子类于彼之元类，如示于图5.3.
彼并类及元类承为彼FinancialHistory程被示于图5.4.
若吾估彼信
FinancialHistory initialBalance:350

p84
彼寻为彼应配initialBalance:始于FinancialHistory class（即FinancialHistory之元类),即，于彼类法于FinancialHistory.
一法为彼择被获于此。
彼法构由两讯：
1.送super讯new.
2.送1值果以讯setInitialBalance:0.

彼寻为new始于FinancialHistory class之超类，即，于Object class.
一法未获于彼，故彼寻续其父链至Class.
彼讯择new被获于Class，且一陋法被行。
彼果乃一未初例于FinancialHistory.此例立被送讯setInicialBalance:.
彼寻为应始于彼例之类，即，于FinancialHistory(于彼例法）。
一法被获其赋一值至每例变。
彼估于
FinancialHistory new

被行于似路。
彼应配new被获于FinancialHistory class(即，于彼类法于FinancialHistory).
彼余行同于initialBalance:除彼参值至setInitialBalance:.
彼例创法须用super new以免招同法轮。

初类变
彼主用于讯至类除创例乃彼初于类变。
彼施绘变申仅给类变以名，无值。
当一类被创，彼名类变被创，然其有值曰nil。
彼元类贯定一法其初彼类变。
如约，彼类变初法贯联与单讯initialize,属初门。

类变被访由类及其元类。 
彼赋值到类变可被做于类法，而非直由一私讯于例法（如其必于例变）。

彼样DeductibleHistory,此番携一类变其需被初，示如下。
DeductilbeHistory乃一子类于FinancialHistory.
其申一类变，MinimumDeductions.

p85

class name DeductibleHistory
superclass FinancialHistory
instance variable names deductibleExpenditures
class variable names MinimumDeductions
class methods

instance creation
  initialBalance:amount
    |newHistory|
    newHistory:=super initialBalance:amount.
    newHistory initializeDeductions.
    ^newHistory
   
new
  |newHistory|
  newHistory:=super initialBalance:0.
  newHistory initializeDeductions.
  ^newHistory.

class initialization
  initialize
    MinimumDeductions:=2300.

instance methods
transcation recording
  spendDeductible:amount for:reason
    self spend:amount for:reason.
    deductibleExpenditures:=
      deductibleExpenditures+amount.

  spend:amount for:reason deducting:deductibleAmount
    self spend:amount for:reason.
    deductibleExpenditures:=
      deductibleExpenditures+deductibleAmount.

inquiries
  isItemizable
    ^deductibleExpenditures>=MinimumDeductions
  totalDeductions
    ^deductibleExpenditures.

private
  initializeDeductions
    deductibleExpenditures:=0.

此版DeductibleHistory添5例法，一为isItemizable.
彼应于此讯乃trueh或false鉴于足税被积以列税于税报。
彼税法定一低税于2300可被收，故若彼积缺，彼标税应被用。
彼常2300被定由类变MinimumDeductions.
为胜送一例于DeductibleHistory彼讯isItemizable,彼类变MinimumDeductions必被赋其数值。
此被做由送彼类彼讯initialize前任例被创。
DeductibleHistory initialize

此讯仅需被送单，后于被类初讯被首定。
彼变被分有每新例于类。

据上类绘，一新例于DeductibleHistory可被创由送彼类彼讯initialBalance:或new,正如为父类FinancialHistory.
设吾估信
DeductibleHistory initialBalance:100.

彼决于哪法被实跟为估彼信鉴于类/父类链为DeductibleHistory.
彼法为initialBalance:被获于类法属DeductibleHistory.

initialBalacne:amount
  |newHistory|
  newHistory:=super initialBalance:amount.
  newHistory initializeDeductions.
  ^newHistory.

此法明newHistory为一临变。
彼首信于彼法乃一赋至彼临变。
newHistory:=super initialBalance:amount.

彼伪变super指彼收。
彼收乃类DeductibleHistory;其类乃其元类。
彼父类于彼元类乃FinancialHistory之元类。
故吾可获彼法进由视于类法于FinancialHistory.

initialBalance:amount.

p87

吾已踪估于此法。
彼应配new被获于Class.
一新例于彼源收，DeductibleHistory，被创且送彼讯setInitialBalance:.
彼寻为setInitialBalance:始于彼新例法，即，于DeductibleHistory.
其未获。
彼寻续至父类FinancialHistory.
其被获且估。
例变被申于FinancialHistory被赋值。
彼值于彼首信于彼类法配initialBalance:于DeductibleHistory,即，乃一部初新例。
此新例被赋至彼临变newHistory.

newHistory即被送彼讯initializeDeductions.
彼寻始于收newHistory之类；彼类乃DedctibleHistory.
彼法被获。
其赋彼值于四例变为0.

彼三信于彼例创讯返彼新例。

一替路去施彼类DeductibleHistory乃陈下。
于此替类绘，彼例创类法于FinancialHistory未被重施。
而，彼私例法讯setInitialBalance:被覆以释添例变。
class name DeductibleHistory
superclass FinancialHistory
instance variable names deductibleExpenditures
class variable names MinimumDeductions
class methods

class initialization
  initialize
    MinimumDeductions:=2300.

instance methods
transaction recording
  spendDeductible:amount for:reason
    self spend:amount for:reason.
    deductibleExpenditures:=
      deductibleExpenditures+amount.

  spend:amount for:reason deducting:deductibleAmount
    self spend:amount for:reason.
    deductibleExpenditures:=
      deductibleExpenditures+deductibleAmount.

inquiries
  isItemizable
    ^deductibleExpenditures>=MinimumDeductions.

  totalDeductions
    ^deductibleExpenditures.

private
  setInitialBalance:amount
    super setInitialBalance:amount.
    dedcutibleExpenditures:=0.

 使此替类绘为DeductibleHistory,彼估于彼应于initialBalance:于

 DeductibleHistory initialBalance:350.

 乃寻于DeductibleHistory元类为initialBalance:.
 其未获。
 续以寻于父类，FinancialHistory元类。获。
 彼法估条于彼信
 super new setInitialBalance:amount.

 彼法配new被获于Class.
 寻为setInitialBalance:始于彼类于新例，一DeductibleHistory.
 彼法配setInitialBalance:被获于DeductibleHistory.
 彼应于setInitialBalance:于DeductibleHistory乃送彼同讯至super故彼寻为彼法始于FinancialHistory.
 其被获且3例变被赋值。
 彼次信于setInitialBalance:于DeductibleHistory置彼柶变为0.
 彼果于彼源讯乃一全初例于DeductibleHistory.

 法定结
 
p89

定彼实行当一讯被送招寻彼法于类团于彼收。
彼寻始于收类且循彼父类链。
若未获至末父类Object,一错被报。
若彼收乃一类，其类乃一元类。
彼讯至其一类可应被列于施绘于部名“类法“。
若彼收非一类，即彼讯至其其可应被列于其施绘于部名”例法“。

彼伪变self指其收于讯其唤彼行法。
彼寻为一法应至一讯至self始于self之类。
彼伪变super亦指彼讯收。
彼寻为一法应一讯至super始于彼父类于彼类于其彼正行法被获。

此终必绘于微言-80程语。
使此统，人须拥普知于彼统类。
部二给细帐于协绘为每一统类且供样，常由陈彼施绘于统类。
部三现一中程。
前探入细于彼实统类，读者或想略至部三得一感于何其定一大程。

术语结

metaclass 元类 彼类于一类。
Class 类类 一理类于诸类除诸元类。(译注：元类乃Metaclass之例）
'''
```
